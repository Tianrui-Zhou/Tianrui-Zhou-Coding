### **Research 1: How has the annual growth of electric vehicle (EV) registrations evolved in Washington State from 1999 to 2024?**



```python
def plot_ev_growth(ev_pop_df):
    """
    This function takes in a DataFrame containing EV population data, and
    plots then return the annual growth of EV registrations in Washington State.
    """
    ev_pop_df = ev_pop_df[(ev_pop_df["State"] == "WA") & (ev_pop_df["Model Year"] != 2025)]

    ev_pop_sum = ev_pop_df.groupby("Model Year")[["DOL Vehicle ID"]].count()

    # Plotting
    plt.figure(figsize=(10, 5))
    ax = sns.lineplot(data=ev_pop_sum, x=ev_pop_sum.index, y="DOL Vehicle ID", marker="o")

    ax.set_xticks(ev_pop_sum.index)
    ax.set_xticklabels(ev_pop_sum.index, rotation=45)
    ax.set_title("Annual Growth of EV Registrations in Washington State")
    ax.set_xlabel("Year")
    ax.set_ylabel("Number of EV Registrations")

    # Add grid
    ax.grid(True)

    return ax

# Example usage
ax = plot_ev_growth(ev_pop_df)
```

##### **Analysis**
From the graph, we can see that the growth of EV registrations in Washington State has gone through several distinct stages.

Between 1999 and 2010, there was almost no movement — EV registrations stayed flat, with little to no adoption during that period.

Starting around 2011 and continuing through 2016, we see the early growth phase, where registrations began to pick up slowly. While the numbers were still relatively small, there was steady progress, growing from just a few  hundred to a few thousand vehicles per year.

From 2017 to 2021, growth really started to take off. This was the acceleration phase, with annual EV registrations increasing rapidly and reaching about 20000 by 2021.

Then came 2022, which marked a huge spike in growth — registrations jumped to nearly 60,000, likely due to a combination of factors like better EV availability, supportive policies, and growing public interest in electric vehicles.

However, after this peak, the graph shows a sharp decline. In 2023, registrations dropped below 50,000, and it looks like the downward trend continued into 2024.

According to our further research, in the overall trend of growth the sudden decrease from 2023-2024 might be the result of these following two reason:
1. Federal Level: Tax Credit Changes for EVs (2023) In August 2022, President Biden signed the Inflation Reduction Act (IRA). This law changed the rules for federal EV tax credits starting in 2023. Now, to get the tax credit (up to $7,500), the electric car must: Be assembled in North America (USA, Canada, Mexico). Use batteries made from materials sourced in the U.S. or approved countries. Because of these new rules, many EVs no longer qualify for the tax credit, so fewer people might buy EVs in 2023 and 2024.

   Source: [Official IRS page](https://www.irs.gov/zh-hans/credits-deductionscredits-for-new-clean-vehicles-purchased-in-2023-or-after)

2. Washington State: Extra EV Registration Fees (2023) In Washington State, starting in 2023, if you own an electric car, you have to pay extra fees every year when you renew your car registration. Here are the extra costs for EV owners: \$150 basic EV fee. \$75 extra fee for EV-related programs. \$75 for EV charging infrastructure (building more charging stations). Total extra cost per year: \$300 (in addition to normal registration fees). These higher fees might make some people less likely to buy an EV.

  Source: [Official Washington State Department of Transportation page](https://wsdot.wa.gov/construction-planning/transportation-modes/vehicles/clean-vehicles/electric-vehicles)


### Machine Learning Models Analysis on Research Question 1: EV Registration Growth in Washington State

We employed four regression models: Linear Regression, Polynomial Regression, Random Forest, and Decision Tree. The goal was to analyze the trend of EV adoption over time and evaluate the performance of these models in predicting EV registration numbers.

```python
# Filter data
ev_pop_df = ev_pop_df[(ev_pop_df["State"] == "WA") & (ev_pop_df["Model Year"] != 2025)]
ev_pop_sum = ev_pop_df.groupby("Model Year")[["DOL Vehicle ID"]].count()
ev_pop_sum.rename(columns={"DOL Vehicle ID": "EV Population"}, inplace=True)

# Prepare data
X = ev_pop_sum.index.values.reshape(-1, 1)  # Model Year
y = ev_pop_sum['EV Population'].values  # EV registrations

# Pre-transform X_poly
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

def evaluate_model(model, X, y, random_states=100):
    """
    This function takes in a model, features, and target variable,
    and returns the best random state model based on the test score.
    """
    best_score, best_rmse, best_state, best_model = -np.inf, None, None, None

    for random_state in range(1, random_states + 1):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)
        model.fit(X_train, y_train)

        test_score = model.score(X_test, y_test)
        rmse = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))

        if test_score > best_score:
            best_score, best_rmse, best_state, best_model = test_score, rmse, random_state, model

    return best_model, best_state, best_score, best_rmse

# 1. Linear Regression
linear_model = LinearRegression()
best_linear_model, best_linear_state, best_linear_score, best_linear_rmse = evaluate_model(linear_model, X, y)

print(f"Linear Regression - Best Random State: {best_linear_state}")
print(f"Linear Regression - Best Test Score: {best_linear_score}")
print(f"Linear Regression - Best RMSE: {best_linear_rmse}\n")

# 2. Polynomial Regression
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
best_poly_model, best_poly_state, best_poly_score, best_poly_rmse = evaluate_model(poly_model, X_poly, y)

print(f"Polynomial Regression - Best Random State: {best_poly_state}")
print(f"Polynomial Regression - Best Test Score: {best_poly_score}")
print(f"Polynomial Regression - Best RMSE: {best_poly_rmse}\n")

# 3. Random Forest Optimization
rf_param_grid = {
    'n_estimators': [50, 100, 150, 170, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=42)
rf_grid_search = GridSearchCV(rf, rf_param_grid, cv=5, scoring="neg_root_mean_squared_error", n_jobs=-1)
rf_grid_search.fit(X, y)

best_rf_params = rf_grid_search.best_params_
print(f"Best Parameters for Random Forest: {best_rf_params}")

best_rf_model = RandomForestRegressor(random_state=42, **best_rf_params)
best_rf_model, best_rf_state, best_rf_score, best_rf_rmse = evaluate_model(best_rf_model, X, y)

print(f"Random Forest - Best Random State: {best_rf_state}")
print(f"Random Forest - Best Test Score: {best_rf_score}")
print(f"Random Forest - Best RMSE: {best_rf_rmse}\n")

# 4. Decision Tree Optimization
dt_param_grid = {
    'max_depth': [None, 2, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

dt = DecisionTreeRegressor(random_state=42)
dt_grid_search = GridSearchCV(dt, dt_param_grid, cv=5, scoring="neg_root_mean_squared_error", n_jobs=-1)
dt_grid_search.fit(X, y)

best_dt_params = dt_grid_search.best_params_
print(f"Best Parameters for Decision Tree: {best_dt_params}")

best_dt_model = DecisionTreeRegressor(random_state=42, **best_dt_params)
best_dt_model, best_dt_state, best_dt_score, best_dt_rmse = evaluate_model(best_dt_model, X, y)

print(f"Decision Tree - Best Random State: {best_dt_state}")
print(f"Decision Tree - Best Test Score: {best_dt_score}")
print(f"Decision Tree - Best RMSE: {best_dt_rmse}")
```

The graph below compares regression models (Linear, Polynomial, Random Forest, Decision Tree) in predicting EV registration growth in Washington State. Black dots show actual data, while colored lines represent model predictions. Polynomial and tree-based models closely fit the exponential growth trend, outperforming Linear Regression.

```python
# Generate predictions
years = np.arange(X.min(), X.max() + 1).reshape(-1, 1)

linear_pred = best_linear_model.predict(years)
poly_pred = best_poly_model.predict(poly_features.transform(years))
rf_pred = best_rf_model.predict(years)
dt_pred = best_dt_model.predict(years)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label="Actual Data")  # Actual data points
plt.plot(years, linear_pred, label="Linear Regression", linestyle="dashed", color="blue")
plt.plot(years, poly_pred, label="Polynomial Regression", linestyle="solid", color="green")
plt.plot(years, rf_pred, label="Random Forest", linestyle="solid", color="red")
plt.plot(years, dt_pred, label="Decision Tree", linestyle="solid", color="orange")

plt.xlabel("Model Year")
plt.ylabel("EV Population")
plt.title("Comparison of Different Regression Models")
plt.legend()
plt.show()
```

```python
def prepare_ev_charge_station_data(ev_charge_station):
    """
    This function takes in the EV charging station DataFrame and prepares it for plotting
    by converting the latitude and longitude columns into geometry data, it will return a
    GeoDataFrame after converting.
    """
    ev_charge_station = ev_charge_station.loc[:, ["Longitude", "Latitude", "ZIP", "City", "Station Name"]]
    ev_charge_station = gpd.GeoDataFrame(
        ev_charge_station,
        geometry=gpd.points_from_xy(ev_charge_station["Longitude"], ev_charge_station["Latitude"],
                                    crs="EPSG:4326")
    )
    return ev_charge_station


def plot_charging_stations(zipcode_geometry, ev_charge_station):
    """
    This function takes in the ZIP code geometry GeodataFrame and EV charging station GeoDataFrame
    and plots then return on a map that represent the distributiong of stations by ZIP Code.
    """
    fig, ax = plt.subplots(figsize=(12, 6))
    zipcode_geometry.plot(ax=ax, color="#EEE", edgecolor="gray", linewidth=0.5)
    ev_charge_station.plot(ax=ax, markersize=10, color='blue', alpha=0.6, edgecolor='k', linewidth=0.5)

    # Adding titles and labels
    ax.set_title("Public Charging Stations by ZIP Code in 2024", fontsize=16)
    ax.set_xlabel("Longitude", fontsize=12)
    ax.set_ylabel("Latitude", fontsize=12)

    # Adding a legend
    import matplotlib.lines as mlines
    legend_ev = mlines.Line2D([], [], color='blue', marker='o', linestyle='None',markersize=10,
                              label='EV Charging Stations')
    ax.legend(handles=[legend_ev], loc='upper left')

    plt.grid(True)
    plt.show()

    return ax


def prepare_ev_population_by_zip(ev_pop_df, zipcode_geometry):
    """
    This method takes in the EV population DataFrame and ZIP code Geometry Dataframe.
    It merged the DataFrams then return a new GeodataFrame that contains the population
    of EV in each ZIP code area in 2024.
    """
    # Filter data for the year 2024
    ev_pop_2024 = ev_pop_df[ev_pop_df["Model Year"] == 2024]

    # Aggregate EV registrations by ZIP code
    ev_pop_zipcode_2024 = ev_pop_2024.groupby("Postal Code")[["DOL Vehicle ID"]].count()

    # Merge with ZIP code geometry
    ev_pop_zipcode = zipcode_geometry.merge(ev_pop_zipcode_2024, left_on="ZCTA5CE10",
                                            right_on="Postal Code", how="inner")

    # Rename columns
    ev_pop_zipcode.rename(columns={"DOL Vehicle ID": "2024 EV Population"}, inplace=True)

    return ev_pop_zipcode


def plot_ev_population_by_zip(ev_pop_zipcode, zipcode_geometry):
    """
    This method takes in a GeoDataFrame containing EV population data by ZIP code
    and a GeoDataFrame containing ZIP code geometry. It then plots and returns a
    Axes represent the map that show the  distribution of EV population by ZIP code.
    """
    fig, ax = plt.subplots(figsize=(12, 6))

    # Plot ZIP code map
    zipcode_geometry.plot(ax=ax, color="#EEE", edgecolor="grey")

    # Plot EV population distribution
    ev_pop_zipcode.plot(ax=ax, column="2024 EV Population", legend=True)

    ax.set_title("2024 EV Population by ZIP Code", fontsize=16)
    ax.set_xlabel("Longitude", fontsize=12)
    ax.set_ylabel("Latitude", fontsize=12)

    plt.grid(True)

    return ax

# Prepare the data
ev_charge_station = prepare_ev_charge_station_data(ev_charge_station)

# Read ZIP code geometry file
zipcode_geometry = gpd.read_file('wa_washington_zip_codes_geo.min.json')

# Plot the data
ax_ev_charging = plot_charging_stations(zipcode_geometry, ev_charge_station)

# Process data
ev_pop_zipcode = prepare_ev_population_by_zip(ev_pop_df, zipcode_geometry)

# Plot
ax_ev_population = plot_ev_population_by_zip(ev_pop_zipcode, zipcode_geometry)
```

```python
def prepare_ev_charger_for_corr(ev_pop_df, ev_charge_station):
    """
    This method takes in EV population DataFrame and EV charge station GeoDataFrame,
    then return a DataFrame of merged EV population and charging station data by ZIP code.
    """
    ev_pop_2024 = ev_pop_df[ev_pop_df["Model Year"] == 2024]
    ev_pop_and_charger_2024 = ev_pop_2024.merge(ev_charge_station,
                                                left_on="Postal Code",
                                                right_on="ZIP",
                                                how="left")

    ev_pop_and_charger_2024_grouped = ev_pop_and_charger_2024.groupby("ZIP").agg(
    {"DOL Vehicle ID": "count", "Station Name": "nunique"}
    ).rename(columns={"DOL Vehicle ID": "2024 EV Population", "Station Name": "Charging Stations"})

    print(type(ev_charge_station))
    return ev_pop_and_charger_2024_grouped


def plot_ev_charger_correlation(ev_pop_and_charger_2024_grouped):
    """
    This method takes in a DataFrame of merged EV population and charging station data by ZIP code.
    Performs correlation analysis and linear regression, then plots the results.
    """
    # Correlation analysis
    correlation = ev_pop_and_charger_2024_grouped.corr()
    print("Correlation matrix:")
    print(correlation)

    # Linear regression
    X = ev_pop_and_charger_2024_grouped["Charging Stations"].values.reshape(-1, 1)
    y = ev_pop_and_charger_2024_grouped["2024 EV Population"].values

    model = LinearRegression()
    model.fit(X, y)

    # Plotting the regression line
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data points')
    plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression line')
    plt.title("Relationship between EV Registration Rates and Public Charging Stations")
    plt.xlabel("Number of Public Charging Stations")
    plt.ylabel("EV Registration Rates (2024)")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print the regression coefficients
    print(f"Intercept: {model.intercept_}")
    print(f"Coefficient: {model.coef_[0]}")

    return correlation, model

grouped_data = prepare_ev_charger_for_corr(ev_pop_df, ev_charge_station)
display(grouped_data)
correlation_matrix, regression_model = plot_ev_charger_correlation(grouped_data)
```

### **Research 3: How do Washington State’s transportation-related greenhouse gas (GHG) emissions correlate with the growth of EV registrations for 2012 to 2017?**

```python
def clean_ghg_data(ghg_df):
  """
  This method takes in WA green house gas emission DataFrame.
  Then return a new DataFrame of transportation related WA
  green house gas emission that is cleaned.
  """
  ghg_df = ghg_df.loc[:, ~ghg_df.columns.str.endswith("biogenic carbon dioxide (MTCO2e)")]

  # Pivot the dataset to long format
  id_vars = ['Source', 'Sector', 'Subsector', 'City', 'County', 'Local Air Authority']
  value_vars = [col for col in ghg_df.columns if col.startswith("20") and "total emissions (MTCO2e)" in col]
  ghg_df_long = ghg_df.melt(id_vars=id_vars, value_vars=value_vars, var_name='Year', value_name='Emission_Value')

  # Extract the year from the first 4 characters of the 'Year' column
  ghg_df_long['Year'] = ghg_df_long['Year'].str[:4]

  # Filter out Transportation GHG data
  ghg_df_long = ghg_df_long.loc[ghg_df_long["Sector"].str.contains("Transportation", case=True, na=False)
                  | ghg_df_long["Subsector"].str.contains("Transportation", case=True, na=False)]

  return ghg_df_long


def filter_ev_data(ev_pop_df):
  """
  This method takes in the EV population DataFrame, then return the
  EV population sum DataFrame for the years 2012-2017.
  """
  # Filter EV population data for the years 2012-2017
  ev_pop_12to17 = ev_pop_df[(ev_pop_df['Model Year'] >= 2012) & (ev_pop_df['Model Year'] <= 2017)]

  # Group by year and sum up the total EV registrations for each year
  ev_pop_yearly_sum = ev_pop_12to17.groupby('Model Year')[['DOL Vehicle ID']].count()
  ev_pop_yearly_sum.rename(columns={'DOL Vehicle ID': 'Total EV Registrations'}, inplace=True)
  ev_pop_yearly_sum.index = ev_pop_yearly_sum.index.astype(str)

  return ev_pop_yearly_sum


def plot_ghg_and_ev_registration_relation(ghg_yearly_sum, ev_pop_yearly_sum):
  """
  This method takes in a DataFrame of WA transportation green house gas emission
  and a DataFrame of EV registrations through year.
  It then return the Axes containing the plot that represent correlation between
  EV Registrations and GHG Emissions (2012-2017).
  """
  # Merge the EV population data with the GHG emissions data
  merged_data = pd.merge(ev_pop_yearly_sum, ghg_yearly_sum, left_index=True, right_index=True)

  # Plotting
  fig, ax1 = plt.subplots(figsize=(10, 6))

  color = 'tab:blue'
  ax1.set_xlabel('Year')
  ax1.set_ylabel('Total EV Registrations', color=color)
  ax1.plot(merged_data.index, merged_data['Total EV Registrations'], color=color, marker='o')
  ax1.tick_params(axis='y', labelcolor=color)

  ax2 = ax1.twinx()
  color = 'tab:red'
  ax2.set_ylabel('Total Emissions (MTCO2e)', color=color)
  ax2.plot(merged_data.index, merged_data['Total Emissions'], color=color, marker='x')
  ax2.tick_params(axis='y', labelcolor=color)

  fig.tight_layout()
  plt.title('Correlation between EV Registrations and GHG Emissions (2012-2017)')
  plt.grid(True)
  return ax1


# Clean the GHG data
cleaned_ghg = clean_ghg_data(ghg_df)
print("Reshaped WA GHG Reporting Dataset:")
display(cleaned_ghg)

# Filter the EV data
ev_pop_yearly_sum = filter_ev_data(ev_pop_df)

# Group by year and sum up the total emissions for each year
ghg_yearly_sum = cleaned_ghg.groupby('Year')[['Emission_Value']].sum()
ghg_yearly_sum.rename(columns={'Emission_Value': 'Total Emissions'}, inplace=True)

# Plot the correlation
plot_ghg_and_ev_registration_relation(ghg_yearly_sum, ev_pop_yearly_sum)
```

##### **Analysis**
Based on 2012-2017 data and correlation plot, there is a weak negative correlation between EV registrations and Washington State's transportation-related greenhouse gas (GHG) emissions. As EV registrations have risen significantly, total transportation-related GHG emissions have only declined slightly. This suggests that while the growing popularity of electric vehicles is a positive trend, its impact in mitigating statewide transportation emissions is limited at best in the near term, possibly due to the continued dominance of gasoline vehicles and the overall increase in transportation demand. Additionally, factors such as insufficient public transit options, and the carbon intensity of the electricity used to charge EVs may also contribute to sustained emission levels. Thus, further EV adoption, paired with supporting policies such as decommissioning combustion engine vehicles and expanding EV infrastructure, may be needed in order to achieve more emission reductions.

**Note:** Although the dual-axis plot attempts to provide a better visualization comparison of EV registrations and transport-based GHG emissions, it can be misleading due to the vast difference in the magnitude of the two datasets. So be careful that GHG emissions are on a significantly larger scale and vary only slightly over time, while EV registrations show a rapid increase.

### Who Benefits?
1. **Policy Makers**: The analysis provides insights into the relationship between EV adoption, charging infrastructure, and greenhouse gas emissions. This information can help in formulating policies to promote EV adoption and reduce emissions.
2. **EV Manufacturers**: The EV automakers can adjust marketing strategies based on areas with growing EV infrastructure, focusing on locations where more charging stations may encourage adoption.

### Who Might Be Excluded or Harmed?
1. **Rural Communities**: The analysis may show that rural areas have fewer charging stations and lower EV adoption rates, potentially leading to less investment in these areas.
2. **Low-Income Populations**: If the analysis focuses on areas with higher income levels where EV adoption is more prevalent, it may overlook the needs of low-income populations who may not afford EVs or have access to charging infrastructure.
3. **Non-EV Users**: The focus on EVs might exclude the needs and preferences of individuals who rely on other forms of transportation, such as public transit or traditional gasoline vehicles.

### Impact of Data Setting
1. **Missing or Incomplete Data**: The dataset might lack private charging station data, at-home charging usage, or incentives that drive EV adoption. This could lead to an underestimation of the total number of charging stations.

### Limitations of the Analysis
1. **Time Sensitivity**: EV adoption is rapidly evolving, and past trends may not accurately predict future behavior. New technological advancements (e.g., improved battery life) or sudden policy changes (e.g., EV tax credits) could significantly shift adoption patterns.
2. **Geographical Scope**: The analysis focuses on Washington State, and the findings may not be generalizable to other regions with different demographics, infrastructure, and policies. Caution should be exercised when applying these conclusions to other areas.
3. **Correlation Does Not Imply Causation**: While the analysis may show correlations between variables such as EV adoption and charging infrastructure, it does not establish causation. Other factors, such as income levels, government incentives, gas prices, and public transportation availability, may also influence these relationships.

### Recommendations for Use
1. **Policy Formulation**: This analysis serves as an initial exploration for the policy makers and should not be considered definitive proof of causality. Future studies could include controlled experiments or additional socioeconomic variables to enhance the conclusions.
2. **Equity Considerations**: Addressing infrastructure gaps in rural and low-income areas is essential to ensure equitable access to clean transportation and avoid exacerbating existing disparities.
3. **Further Research**: Researchers should build on this analysis by incorporating more recent data, expanding the geographical scope, and exploring additional factors influencing EV adoption and emissions reduction.
